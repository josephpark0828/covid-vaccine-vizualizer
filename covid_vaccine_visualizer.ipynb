```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter, WeekdayLocator
from datetime import timedelta
%matplotlib inline
```

```python
df = pd.read_csv('https://covid.ourworldindata.org/data/owid-covid-data.csv', usecols= ['date', 'location', 'total_vaccinations_per_hundred'], parse_dates=['date'])

countries = ['United States', 'South Korea', 'United Kingdom', 'Germany']

df = df[df['location'].isin(countries)]

pivot = pd.pivot_table(
    data=df,
    index='date',
    columns='location',
    values='total_vaccinations_per_hundred',
    aggfunc='mean',
)

fig, ax = plt.subplots(figsize=(12,8))

for country in countries:
  country_color="red"
  if (country == "South Korea"):
    country_color="blue"
  elif (country == "United Kingdom"):
    country_color="green"
  elif (country == "Germany"):
    country_color="black"

  ax.plot(
      pivot.index,
      pivot[country],
      color=country_color,
      alpha=1.0
  )

plt.show()
```
